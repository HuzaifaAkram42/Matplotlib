## Types of Data
#### Numerical Data
- Definition: Numerical data that can be measured and quantified.
    - Subtypes:
    - Discrete Data: Countable data with a finite number of values (e.g., the number of students in a class).
    - Continuous Data: Data that can take any value within a range (e.g., height, weight, temperature).a

#### Categorical Data
- Definition: Data that can be divided into specific categories.
    - Subtypes:
    - Binary Data: Data with only two possible categories (e.g., yes/no, true/false).
    - Multi-class Data: Data with more than two categories (e.g., types of vehicles: car, truck, bike).

## 2D Line plot
A 2D line plot is a type of chart used to display data points connected by straight lines in a two-dimensional space, typically used to visualize trends over time or relationships between two variables.

- Bivariate Analysis: 
Bivariate analysis involves the comparison of two variables to understand the relationship between them.
- Key Components:
    - X-axis: Represents the independent variable (e.g., time, categories).
    - Y-axis: Represents the dependent variable (e.g., values, measurements).
    - Data Points: Each point on the plot corresponds to an (X, Y) pair.
    - Line: The points are connected by lines, showing how the dependent variable changes with the independent variable.
- Usage:
  - Commonly used in time-series analysis, where the X-axis is time.
  - Helpful for identifying trends, patterns, and correlations between variables.
- Example:
    - A line plot showing monthly sales revenue over a year, with months on the X-axis and revenue on the Y-axis.

#### categorical -> numerical
- This analysis examines the relationship between a categorical variable (e.g., categories, labels) and a numerical variable (e.g., measurements, counts).
- Example:
Suppose you have sales data for different product categories (e.g., Electronics, Clothing, Groceries) over time. Here, the categorical variable is the product category, and the numerical variable is sales revenue.

```python
data = {
    'Date': pd.date_range(start='2024-01-01', periods=12, freq='M'),
    'Electronics': [200, 220, 210, 215, 230, 250, 240, 235, 245, 255, 260, 270],
    'Clothing': [150, 160, 155, 165, 170, 175, 180, 185, 190, 195, 200, 205],
    'Groceries': [100, 105, 110, 115, 120, 125, 130, 135, 140, 145, 150, 155]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Electronics'], marker='o', label='Electronics')
plt.plot(df['Date'], df['Clothing'], marker='o', label='Clothing')
plt.plot(df['Date'], df['Groceries'], marker='o', label='Groceries')

plt.title('Monthly Sales by Category')
plt.xlabel('Date')
plt.ylabel('Sales Revenue')
plt.legend()
plt.grid(True)
plt.show()
```

#### numerical -> numerical
- This analysis examines the relationship between two numerical variables.
- Example:
    Imagine you have a dataset containing temperature and ice cream sales over time. Both temperature and sales are numerical variables, and the goal is to analyze how changes in temperature influence sales over time.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Date': pd.date_range(start='2024-01-01', periods=12, freq='M'),
    'Temperature': [30, 32, 35, 33, 31, 30, 29, 28, 27, 26, 25, 24],
    'IceCreamSales': [200, 210, 220, 230, 225, 220, 215, 210, 200, 190, 180, 170]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(df['Date'], df['Temperature'], color='blue', marker='o')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (C)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()  # Create a second y-axis
ax2.plot(df['Date'], df['IceCreamSales'], color='red', marker='o')
ax2.set_ylabel('Ice Cream Sales', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Temperature vs Ice Cream Sales Over Time')
plt.grid(True)
plt.show()
```

## Scatter plots:

- Key Features:
    - X-Axis & Y-Axis: Represent the two variables being compared.
    - Data Points: Each point represents an observation, plotted according to its values on the X and Y axes.
    - Correlation: Scatter plots help identify correlations or relationships between the variables (e.g., positive, negative, or no correlation).

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

## Bar Chart
Bar charts are used to display and compare the quantity or frequency of categorical data. Each bar represents a category, with the length or height of the bar indicating the value or count for that category.

- Key Features:
    - X-Axis: Represents categories or discrete groups.
    - Y-Axis: Represents values or frequencies.
    - Bars: The height or length of each bar corresponds to the value of the category it represents.
- Usage:
    - Comparing Categories: Useful for comparing different categories or groups.
    - Displaying Frequencies: Effective for showing the count of occurrences in each category.

```python
# Sample data
categories = ['Red', 'Blue', 'Green', 'Yellow' , 'Pink']
values = [10, 15, 7, 12, 7]
# Bar chart
colors = ['Red', 'Blue', 'Green', 'Yellow' , 'Pink']
plt.bar(categories, values, color=colors, alpha=0.7)
# Adding labels and title
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Favorite Color')
# Display the plot
plt.show()
```

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# Horizental bar Chart is create plt.barh
plt.barh(categories, values, color=colors, alpha=0.7)
```

## Histogram
A histogram is a type of plot used to represent the distribution of numerical data by dividing the data into bins (intervals) and showing the frequency or count of data points within each bin.
    
- Key Features:
    - Bins: Intervals into which the data is divided. Each bin represents a range of values.
    - Bars: The height of each bar indicates the frequency or count of data points falling within the corresponding bin.
- Usage:
    - Distribution Analysis: To understand the distribution and spread of a dataset.
    - Frequency Distribution: To visualize how often values fall within certain ranges.

```python
# Sample data
data = np.random.randn(1000)  # 1000 random numbers from a normal distribution
# Histogram
plt.hist(data, bins=30, edgecolor='Red')
# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
# Display the plot
plt.show()
```

## Pie Chart
A pie chart is a circular statistical graphic that is divided into slices to illustrate numerical proportions. Each slice represents a category's contribution to the whole.

- Key Features:
    - Slices: Represent different categories or groups.
    - Proportions: The size of each slice is proportional to the percentage or fraction of the total for that category.
    - Labels: Each slice can be labeled to identify the category it represents.
- Usage:
    - Proportional Representation: To show the relative sizes of parts to the whole.
    - Comparing Parts: Useful for displaying the composition of a single data series.

- Univariate/Bivariate Analysis- 
Categorical vs numerica
- Use - To find contibution on a standard scale

```python
# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]  # Corresponding sizes for each category
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']  # Optional: colors for the slices
# Pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
# Adding title
plt.title('Pie Chart Example')
# Display the plot
plt.show()
```
