## Colored Scatterplot

```python
iris = pd.read_csv("Datasets\\iris.csv")
iris.head()
```

```python
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-virginica':1, 'Iris-versicolor':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
```

```python
# for coloring we use c parameter and pass the species value
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Lenght")
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'], cmap='jet')
plt.xlabel("Sepal Length")
plt.ylabel("Petal Lenght")
plt.colorbar()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'], cmap='jet',alpha=0.5)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Lenght")
plt.colorbar()
```

```python
plt.figure(figsize=(16,8))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'], cmap='jet',alpha=0.5)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Lenght")
plt.colorbar()
```

```python
batter = pd.read_csv("Datasets\\batter.csv")
batter.head()
```

```python
batter.shape
```

```python
selected_batter = batter.head(100).sample(25,random_state=5)
selected_batter
```

```python
plt.figure(figsize=(15,8))
plt.scatter(selected_batter['avg'],selected_batter['strike_rate'], s=selected_batter['runs'])
plt.xlabel("Batter Average")
plt.ylabel("Batter Strike Rate")
for i in range(selected_batter.shape[0]):
    plt.text(selected_batter['avg'].values[i],selected_batter['strike_rate'].values[i],selected_batter['batter'].values[i])
```

## Horizental and Vertical Lines

```python
plt.figure(figsize=(15,8))
plt.scatter(selected_batter['avg'],selected_batter['strike_rate'], s=selected_batter['runs'])
plt.xlabel("Batter Average")
plt.ylabel("Batter Strike Rate")
for i in range(selected_batter.shape[0]):
    plt.text(selected_batter['avg'].values[i],selected_batter['strike_rate'].values[i],selected_batter['batter'].values[i])

plt.axhline(130,color='red')
plt.axvline(30,color='blue')
plt.axhline(140,color='green')
```

```python
plt.figure(figsize=(15,6))
plt.scatter(batter['avg'],batter['strike_rate'])
plt.title("Average vs Strike Rate")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax.scatter(batter['avg'],batter['strike_rate'],color='green',marker='>')
ax.set_title("Average vs Strike Rate")
ax.set_xlabel("Average")
ax.set_ylabel("Strike Rate")
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1)
ax[0].scatter(batter['avg'],batter['strike_rate'])
ax[1].scatter(batter['avg'],batter['runs'])
ax[0].set_title("Average Vs Strike Rate")
ax[0].set_ylabel("Strike Rate")
ax[1].set_title("Average Vs Runs")
ax[1].set_ylabel("Runs")
ax[1].set_xlabel("Average")
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(batter['avg'],batter['strike_rate'],color='green')
ax[1].scatter(batter['avg'],batter['runs'],color='blue')
ax[0].set_title("Average Vs Strike Rate")
ax[0].set_ylabel("Strike Rate")
ax[1].set_title("Average Vs Runs")
ax[1].set_ylabel("Runs")
ax[1].set_xlabel("Average")
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].scatter(batter['avg'],batter['strike_rate'],color='green')
ax[0,1].scatter(batter['avg'],batter['runs'],color='blue')
ax[1,0].hist(batter['avg'],color='purple')
ax[1,1].hist(batter['runs'],color='orange')

ax[0,0].set_title("Average Vs Strike Rate")
ax[0,1].set_title("Average Vs Runs")
ax[1,0].set_title("Average")
ax[1,1].set_title("Runs")
```

```python
fig = plt.figure(figsize=(16,6))

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batter['avg'],batter['strike_rate'],color='green')

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(batter['avg'],batter['runs'],color='green')

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batter['avg'],color='purple')

ax3 = fig.add_subplot(2,2,4)
ax3.hist(batter['runs'],color='orange')
fig.show()
```

## 3D Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
x
```

```python
xx , yy = np.meshgrid(x,y)
```

```python
xx
```

```python
xx.shape
```

```python
z = xx**2 + yy**2
fig = plt.figure(figsize=(10,5))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(10,5))
ax = plt.subplot(projection='3d')
p2 = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p2)
```

## Heatmap

```python
delvries = pd.read_csv("Datasets\\IPL_Ball_by_Ball_2008_2022.csv")
delvries.head()
```

```python
delvery = delvries[(delvries['ballnumber'].isin([0,1,2,3,4,5])) & (delvries['batsman_run'] == 6)]
delvery
```

```python
grid = delvery.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
grid
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delvries['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.colorbar()
```
